# Project Motivation

In the business world, advertising your products with promotional offers is a typical strategy. In order to succeed its advertising campaign, the company must target the right customers.  
Since a promotional offer is **expensive** and can result in lost profits for the company, it must be **well studied**.  
In this project, we want to find 10 regions that have similar consumption habits to "Saint Petersburg" where a promotional offer on wine was successful. Our ultimate goal is to **maximize sales**.

# Success criteria

We assume that a region's consumption habits would be favorable to the success of the promotion with a maximum of sales if this region has both criteria :
- consumption distribution similar to Saint Petersbourg
- high wine consumption during the last year (2016)  
  
We will choose the 10 regions that best fit these criteria

## Saint Petersburg's consumption distribution

We assume that the campaign will be successful in regions where **wine consumption is similar** to Saint Petersburg over time. Thus we will be treating data as time series and look for the most similar time series to Saint Petersburg.

```python
%matplotlib inline
# plotting consumption distribution of 'Saint Petersburg'
products=["beer","wine","vodka","champagne","brandy"]
sp_df = df[df["region"]=="Saint Petersburg"].set_index("year",drop=True)[products]
sp_df.plot.line()
plt.ylabel('consumption in liters')
plt.title('Alcohol consumption in Saint Petersburg')
plt.savefig('alc_consumption_in_S_P.png')
plt.show()
```

We can not really see the distrubutions. Let's scale by MaxAbsScaler(2)

## Focus on wine
Let's visualize the correlation matrix of mean consumption per year per each product

```python
# Creating a table of mean consumption per year per each product
consumption_grouped_by_year = df.groupby(["year"]).mean()
```

```python
# Generate a mask for the upper triangle 
mask = np.triu(np.ones_like(df.groupby(["year"]).mean().corr(), dtype=bool))
# Add the mask to the heatmap
sns.heatmap(df.groupby(["year"]).mean().corr(), mask=mask, center=0, linewidths=1, annot=True, fmt=".2f")
plt.savefig("correlation")
plt.show()
```

- Since the purchase of the different products seem to be highly correlated with wine (except for vodka) We will keep the analysis simple and only focus on the wine time series
- We will create a new dataframe in wich each sample is a different region's wine consumption from 1998 to 2016.
- We  will use cosine (1) as metric of distance  instead of euclidean because we want to focus on the form of the data series.


```python
# Sorting values and setting region as index
df = df.sort_values(['region','year'])
df = df.set_index('region') 
```

```python
# creating the dataframe: data
subset_to_delete = ["year","beer","vodka","champagne","brandy"]
df_list = []
# creating list of dataframes for each year
for i in range(1998,2017):
    df_list.append(df[df['year'] == i].drop(subset_to_delete, axis=1)) 
    
# concatenating in a new dataframe: data
data = pd.concat(df_list,axis=1)
# renaming the columns
list_of_new_columns =[str(year) for year in range(1998,2017)]
data.columns = list_of_new_columns

#checking the new dataset: data
data.head(2)
```

So we got a dataframe of wine consumption from 1998 to 2016 for each region.

## Data Preparation
- To make all regions comparable and only focus on the evolution, it is logical to scale each region by the maximum of its history.

```python
# Scaling data 
data_scaled = data.div (data.max(axis= 1),axis = 0).values
```

## TSNE Visualisation 
How many clusters ?

```python
# Creating the list of all regions:all_regions
all_regions =list(data.index)
# Import TSNE
from sklearn.manifold import TSNE
# Create a TSNE instance: model
model = TSNE(learning_rate=150)
# Apply fit_transform to samples: tsne_features
tsne_features = model.fit_transform(data_scaled)
# Select the 0th feature: xs
xs = tsne_features[:,0]
# Select the 1st feature: ys
ys = tsne_features[:,1]
# Scatter plot, coloring by variety_numbers
plt.scatter(xs ,ys,alpha=0.5)
plt.scatter( tsne_features[58,0] , tsne_features[58,1] ,marker ='s',c='r')
for x, y, company in zip(xs, ys, all_regions):
    plt.annotate(company, (x, y), fontsize=5, alpha=0.75)
plt.show()
```

No clear separation between the different points. Let's try "Elbow method".


## Elbow method

```python
# (Elbow method)
from sklearn.cluster import KMeans
L = []
for i in range(2,15) :
    model=KMeans(n_clusters= i,init='k-means++',random_state =42) 
    # init=k-means++ (3)
    model.fit(data_scaled)
    L.append(model.inertia_)
    
plt.plot(range(2,15) , L,'rp--');
plt.xlabel("Number of clusters");
plt.ylabel("Inertia");
plt.title('The Elbow method');
plt.grid()
```

The Elbow method does not work 


# Model Development

Since the number of data points is not huge, we opt for a **hierarchical clustering**, where we will draw the dendogram and cut the linkage tree at the threshold that best separates Saint Petersburg from the rest.  
We are dealing with time series and we will be using cosine similarity/distance.
We use cosine instead of euclidean because we want to focus on the form of the data series.

## Method 1: Focusing only on similar regions to St. Petersburg 

```python
# Fitting the data into a hierarchical clustering algorithm
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
mergings = linkage(data_scaled, method ='average',metric='cosine')
# plotting dendrogram
dn = dendrogram(mergings, labels=all_regions, leaf_rotation=90, leaf_font_size=6)
plt.show()
```

Looking at the dendrogram we can see that we can cut it at around 0.04 to obtrain a cluster containing 'Saint Petersburg' and  10 other regions


```python
# Assign cluster labels to each row of data
labels = data['label_1'] = fcluster(mergings,0.04, criterion='distance')
```

```python
# finding the label of the cluster that contains "Saint Petersburg"
SP_cluster_label = data.loc["Saint Petersburg"]["label_1"]
```

```python
# printing the regions in the cluster that contains "Saint Petersburg"
list_1 = list(data[data['label_1']==SP_cluster_label].index) 
len(list_1)
```

```python
# Looking for wine consumption distribution of list_1
for region in list_1:
    plt.plot(data.drop('label_1', axis=1).loc[region])
plt.plot(data.drop('label_1', axis=1).loc['Saint Petersburg'],linewidth=3,c='k',linestyle='--',label='St.Petersburg')
plt.xticks(rotation=45)
plt.xlabel('year')
plt.ylabel('consumption in liters')
plt.legend()
plt.show()
```

We can see some similarity in these distributions

## Analysing 2016 consumption
let's look at wine consumption in the last year (2016) in these regions, then compare it to the overall top 11 wine consumption:

## Method 2: Focus in similar regions to St. Petersburg and Maximum wine consumption

Here we use method = 'complete' , to make the clusters more condensed

```python
# Fitting the data into a hierarchical clustering algorithm
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
mergings = linkage(data_scaled, method ='complete', metric='cosine')
```

```python
# plotting the dendrogram
dn = dendrogram(mergings, labels=all_regions, leaf_rotation=90, leaf_font_size=6)
plt.title('hierarchical clustering')
plt.savefig('hirerachy')
plt.show()
```

```python
# Assigning cluster labels to each row of data
labels = data['label_2'] = fcluster(mergings,0.04, criterion='distance')
# finding 'Saint Petersburg' label: Sp_label_2 
Sp_label_2 = data.loc["Saint Petersburg"]["label_2"]
# printing the regions in the cluster that contains "Saint Petersburg"
list_2 = list(data[data['label_2']==Sp_label_2].index)
len(list_2)
```

In this case we end up with a cluster of 19 regions : we have 18 regions similar to 'Saint Petersburg'. 
This is not so bad since it allows us to take into account the mean consumption in each regions that we ignored by normalizing. 
For this reason We decide to choose the regions with the most max wine sales in their 2016.

```python
list_2
```

```python
list_2_sorted = data.loc[list_2]['2016'].sort_values(ascending=False)
```

```python

list_2_sorted.plot.barh(color =['g','g','g','g','g','g','g','g','g','g','r','b','b','b','b','b','b','b','b'],align='center')
plt.ylabel('wine consumption')
plt.savefig('finallista')
plt.title('')
plt.show()
```

```python
# Sorting in descending order according to wine consumption in 2016 and taking top 10:
final_list=list(data.loc[list_2]['2016'].sort_values(ascending=False)[:10].index)
# The final list
final_list
```

We notice that we don't obtain the same list as the previous method.

```python
# Looking for the distribution 
for region in final_list:
    plt.plot(data.drop(['label_1','label_2'], axis=1).loc[region])
plt.plot(data.drop(['label_1','label_2'], axis=1).loc['Saint Petersburg'],linewidth=3,c='k',linestyle='--',label='St.Petersburg')
plt.xticks(rotation=45)
plt.ylabel('consumption in liters')
plt.title('10 regions dist. similar to St.Petersburg')
plt.legend()
plt.savefig('list_2_distribution')
plt.show()
```

These regions have more similar distributions compared to the previous method!

Finally to visualize this similarity, let's create a general view of wine consumption distribution in all regions

```python
# Creating non_selected_set = list of all regins without St.Petersburg - final_list
non_selected_set = set(all_regions).difference('Saint Petersburg').difference(set(final_list))
```

```python
# Looking for the distributions 
for region in final_list:
    plt.plot(data.drop(['label_1','label_2'], axis=1).loc[region],c='k',linewidth=0.8,alpha=0.9)
for region in non_selected_set:
    plt.plot(data.drop(['label_1','label_2'], axis=1).loc[region],linewidth=0.5,alpha=0.5)
plt.plot(data.drop(['label_1','label_2'], axis=1).loc['Saint Petersburg'],linewidth=3,c='k',linestyle='--',label='St.Petersburg')
plt.xticks(rotation=45)
plt.title('Wine consumption all regions')
plt.ylabel('consumption in liters')
plt.legend()
plt.savefig('results')
plt.show()
```

As we can see the regions of our final_list (black lines) have distributions more **similar** to St.Petersburg and a **hight** mean consumption compared to the other regions (coored lines).

For future works:  
- We recommend to add the **mouthly consumption** of each product to the dataset: it can be usefull for choosing the best time for a promotion and can help the algorithm with more details about the time series.
